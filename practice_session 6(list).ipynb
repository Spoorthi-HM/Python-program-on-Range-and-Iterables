# packing and unpacking

```python
#unpacking
L1=[25,56,71,36,78,39]
a,b,c,d,e,f = L1
print(a,b,c,d,e,f)
#packing
L2=[a+2,b-1,c+5,d-2,e+3,f-4]
print(L2)
```

```python
L3=['a','b','c','d']
p,q,r,s=L3
print(p,q,r,s)
L4=[1,2,3,4,5]
a,b,c,d,e=L4
L4=[a+6,b-2,c-1,d+5,e+6]
print(L4)
L6=[1.4,2.6,3.8,5.9,4.4]
m,n,o,p,q=L6
print(m,n,o,p,q)
L6=[m-1,n+5,o-3,p+4,q+2]
print(L6)
```

```python
L5=["coffee","bread","tea","sunny"]
d,e,f,g=L5
print(d,e,f,g)
L5=[d+"-tea",e+"-jam",f+"-cup",g+"-morning"]
print(L5)
```

```python
L7=["morning","afternoon","evening","night"]
r,s,t,u=L7
print(r,s,t,u)
L7=[r+"-coffee",s+"-juice",t+"-tea",u+"-milk"]
print(L7)
```
