# list methods

```python
# removing the values from the list
L2=['a','b','c','d','e','f']
L2.remove(L2[3])
print(L2)
# add the elements in the end to the list
L3=["Alice","Bob","John"]
L3.append("Ronny")
print(L3)
# delete the values in the list
L4=['P','Q','R','S']
del L4[0]
print(L4)
#Editing the elements
L5=[1,2,3,4,5]
L5[0]=6
L5[1]=20
L5[2]=8
L5[3]=12
L5[4]=2
print(L5)
# add the values to the list
L6=["cat","dog","rat"]
L6.insert(2,'mouse')
print(L6)
```
