# Slicing operators

```python
l1=[12,34,56,34,28,67,45,55,64]
print(l1[0:7:3])
```

```python
l2=[23,35,18,56,78,13,10,65,34,67]
print(l2[1:5:2])
print(l2[0:5])
print(l2[2:8])
print(l2[:4])
print(l2[5:])
print(l2[-3:])
print(l2[:-2])
print(l2[::-2])
print(l2[1::2])
print(l2[::-1])
print(l2[8:2:-1])
```
